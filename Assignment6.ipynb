```python
!pip install requests
import requests
target_url = 'http://www.gutenberg.org/files/2265/2265.txt'
response = requests.get(target_url)
data = response.text
```

```python
from nltk.corpus import stopwords
from textblob import TextBlob
blob = TextBlob(data)
stop_words = stopwords.words('english')
freq = blob.word_counts.items()
freq = [freq for freq in freq if freq[0] not in stop_words]
```

```python
from operator import itemgetter
sorted_items = sorted(freq, key=itemgetter(1), reverse=True)
```

```python
top_20 = sorted_items[1:21]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame(top_20, columns=['word', 'count'])
df
```

```python
axes = df.plot.bar(x='word', y='count', legend=False)
plt.gcf().tight_layout()
```
